```python
user=df
```

```python
user.groupby('occupation')['age'].mean().reset_index()
#user.groupby('occupation')['gender'].value_counts().reset_index()
```

```python
m=user[(user["gender"]=='M')]
```

```python
m.groupby('occupation')['gender'].count().sort_values(ascending=False).reset_index()
#t.gender.count()
#df.groupby('occupation')['gender'].value_counts().unstack(fill_values=0).reset_index()
#count.columns=["",""]
#count
```

```python
count=user.groupby('occupation')['gender'].value_counts().unstack(fill_value=0).reset_index()
count.columns=["occupation","F",'M']
count['Total']=count['F']+count['M']
count.sort_values(by='Total',ascending=False)
```

```python
count['male%']=(count['M']*100)/count['Total']
```

```python
count['female%']=(count['F']*100)/count['Total']
```

```python
count.sort_values(by='Total',ascending=False)
```
