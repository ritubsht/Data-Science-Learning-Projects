```python
import pandas as pd
df=pd.read_excel("crop_data.xlsx")
```

```python
df
```

```python
df.drop(columns="Unnamed: 0", inplace=True)
```

```python
df
```

```python
df["Season"]=df["Season"].str.strip()
```

```python
df
```

```python
df["Season"].unique()
```

```python
#Which of these crops are produced during the Summer season? 
Summer=df[df["Season"]=="Summer"]
```

```python
Summer["Crop"].unique()
```

```python
#During which year did Haryana have the highest crop production?
hary=df[df["State_Name"]=="Haryana"]
hary
hary.groupby("Crop_Year")["Production"].sum().sort_values(ascending=False)
```

```python
#The maximum and minimum area for production were in the years?
df.groupby("Crop_Year")["Area"].agg(['min', 'max'])
```

```python
#Which state in India had the second lowest crop production? (overall, for all years)
df.groupby(["State_Name"])["Production"].sum().sort_values(ascending=False).reset_index().iloc[-2]
```

```python
#What were the top three produced crops in the year 2012?
a=df[df["Crop_Year"]==2012]
a.groupby("Crop")["Production"].sum().sort_values(ascending=False).head(3)
```

```python
#What is the standard deviation for Area of production?
df['Area'].std()
```

```python
#Which is the crop that gave the highest production to the state of Andhra Pradesh?
AP=df[df["State_Name"]=="Andhra Pradesh"]
AP
AP.groupby("Crop")["Production"].sum().sort_values(ascending=False).head(1)
```

```python
df.groupby("Season")["Production"].sum().sort_values(ascending=False)
```

```python
#Which state has the lowest area of production?
df.groupby("State_Name")["Area"].sum().sort_values(ascending=False).tail(1)
```

```python
#What is the mean for the area of production?
df.Area.mean()
```

```python
#What is the correlation coefficient between Area and Production?
df["Area"].corr(df["Production"])
```

```python
#Which is the only crop that has the highest production during autumn?
autumn=df[df["Season"]=="Autumn"]
autumn
autumn.groupby("Crop")["Production"].sum().sort_values(ascending=False).head(1)
```

```python
#Which is the only crop that has the highest production during summer?
summer=df[df["Season"]=="Summer"]
summer
summer.groupby("Crop")["Production"].sum().sort_values(ascending=False).head(1)
```

```python
#Which is the only crop that has the highest production during winter?
winter=df[df["Season"]=="Winter"]
winter
winter.groupby("Crop")["Production"].sum().sort_values(ascending=False).head(5).plot(kind="pie");
```
