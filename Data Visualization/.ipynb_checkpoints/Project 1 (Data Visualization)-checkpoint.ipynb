```python
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data
data = {'Category': ['A', 'B', 'C'], 'Values': [10, 15, 7]}
sns.set_theme(style="dark")

# Create a barplot
ax = sns.barplot(x='Category', y='Values', data=data, palette='pastel')

# Add hatches and customize their colors
hatches = ['/', '\\', 'x']  # Define hatch patterns
colors = ['red', 'blue', 'green']  # Define custom colors

for bar, hatch, color in zip(ax.patches, hatches, colors):
    bar.set_hatch(hatch)  # Apply hatch pattern
    bar.set_edgecolor(color)  # Set edge color
    bar.set_linewidth(1.5)  # Adjust line width for better visibility

# Show the plot
plt.show()
```

```python
months=["Jan","Feb", "Mar", "Apr", "May", "June","Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
#sales data in thousands
sales_2024=[50,55,53,60,65,70,75,72,68,70,74,80]
sales_2025=[52,58,56,62,70,75,78,76,73,77,79,85]

plt.plot(months,sales_2024,label="sales_2024", color= "blue", marker="o")
plt.plot(months,sales_2025,label="sales_2025", color= "green", marker="s")
plt.xlabel("Month")
plt.ylabel("sales in thousands")
plt.title("Monthly sales trends 2024 vs 2025")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

```

```python
import pandas as pd
months=["Jan","Feb", "Mar", "Apr", "May", "June","Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
#sales data in thousands
sales_2024=[50,55,53,60,65,70,75,72,68,70,74,80]
sales_2025=[52,58,56,62,70,75,78,76,73,77,79,85]

data=pd.DataFrame({"Month":months*2, "Sales":sales_2024 + sales_2025,"Year":['2024']*12 +['2025']*12})
```

```python
data
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(10,6))
sns.scatterplot(data=data, x='Month', y='Sales', hue='Year',marker='o')

plt.xlabel("Month")
plt.ylabel("sales in thousands")
plt.title("Monthly sales trends 2024 vs 2025")

plt.tight_layout()
plt.show()



```
