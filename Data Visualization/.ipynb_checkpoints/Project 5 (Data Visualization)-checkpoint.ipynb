```python
dpt=df.groupby('Department')['Salary_INR'].mean()
dpt
```

```python
dpt=df.groupby('Department')['Salary_INR'].mean()
dpt.plot(x=dpt.index, y=dpt.values, kind ='bar', color='c', legend= True, width =0.3)
plt.grid()
plt.title("Av. salary of dept")
plt.ylabel("Salary")
plt.show()
```

```python
s=df.groupby('Job_Title')['Salary_INR'].mean().sort_values(ascending=False)
s.plot(x=s.index, y=s.values, kind ='bar', color='g', legend= True, width =0.3)
plt.grid()
plt.title("Av. salary wrt Job title")
plt.ylabel("Salary")
plt.show()
```

```python
s2=df[(df['Status']=='Resigned') | (df['Status']=='Terminated')]
s3=s2.groupby('Department')['Status'].count()
s3.plot(x=s3.index, y=s3.values, kind ='bar', color='g', legend= True, width =0.3)
plt.grid()
plt.title("No. of terminated and resigned employee")
plt.ylabel("count")
```
