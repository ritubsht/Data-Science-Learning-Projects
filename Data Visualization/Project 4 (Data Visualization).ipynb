```python
import sklearn
df.dropna(inplace=True)
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, precision_score, recall_score, f1_score

X = df.drop('Satisfied', axis=1)
y = df['Satisfied']

# Splitting the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42,stratify=y)
```

```python
# Identifying column dtypes
categorical_features = X.select_dtypes(include=['object', 'category']).columns
numerical_features = X.select_dtypes(include=np.number).columns
numerical_features = numerical_features.drop(['Ref', 'id'])

# Initializing scaler & encoder
scaler = StandardScaler()
encoder = OneHotEncoder(handle_unknown='ignore')

# Fit and transform the training data
X_train_num_scaled = scaler.fit_transform(X_train[numerical_features])
X_train_cat_encoded = encoder.fit_transform(X_train[categorical_features])

# Only transforming the testing data
X_test_num_scaled = scaler.transform(X_test[numerical_features])
X_test_cat_encoded = encoder.transform(X_test[categorical_features])

```

```python
X_train_processed = np.hstack((X_train_num_scaled, X_train_cat_encoded.toarray()))
X_test_processed = np.hstack((X_test_num_scaled, X_test_cat_encoded.toarray()))
```

```python
# --- 4. Train and Evaluate the Model ---
model = LogisticRegression(random_state=42)

# Train the model on the manually processed training data
model.fit(X_train_processed, y_train)

```

```python
# predictions
y_pred_lr = model.predict(X_test_processed)

# Metrics for Logistic Regression
lr_accuracy = accuracy_score(y_test, y_pred_lr)
lr_precision = precision_score(y_test, y_pred_lr, pos_label='Y')
lr_recall = recall_score(y_test, y_pred_lr, pos_label='Y')
lr_f1 = f1_score(y_test, y_pred_lr, pos_label='Y')
```

```python
print(f"Logistic Regression Accuracy: {lr_accuracy:.4f}")
print(f"Logistic Regression Precision (Positive 'Y'): {lr_precision:.4f}")
print(f"Logistic Regression Recall (Positive 'Y'): {lr_recall:.4f}")
print(f"Logistic Regression F1-Score (Positive 'Y'): {lr_f1:.4f}")
```
