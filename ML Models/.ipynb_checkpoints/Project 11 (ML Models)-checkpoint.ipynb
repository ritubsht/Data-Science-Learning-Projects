```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")

from sklearn.metrics import accuracy_score, classification_report, precision_score, recall_score, f1_score
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
```

```python
data ={"Study_Hour":[2,4,6,8,1,7,3,9,5,6],
       "Attendence":[60,70,80,90,50,85,65,95,75,80],
       "Pass":[0,0,1,1,0,1,0,1,1,1]}
```

```python
df=pd.DataFrame(data)
```

```python
df
```

```python
df.info()
```

```python
df.columns
```

```python
#cr=df.corr()
#annot='Set1 or Set2 or Set3 or Coolwarm or Greys'
```

```python
sns.heatmap(df[['Study_Hour', 'Attendence', 'Pass']].corr(), annot=True, cmap='Set3', fmt=".2f")
plt.title('Correlation between features')
plt.show()
```

```python
X= df[['Study_Hour', 'Attendence']]
y=df['Pass']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3, random_state=42)
```

```python
X_train
```

```python
X_test
```

```python
dt=DecisionTreeClassifier(random_state=42)
dt.fit(X_train, y_train)
```

```python
pred=dt.predict(X_test)
accuracy_score(y_test,pred)
```

```python
plt.figure(figsize=(8,6))
tree.plot_tree(dt,feature_names= X.columns,class_names=["Fail", "Pass"], filled=True)
plt.show()

```
