```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import sklearn
import warnings
warnings.filterwarnings("ignore")
from sklearn.linear_model import LinearRegression
from word2number import w2n
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
df=pd.read_csv('hiring.csv')
```

```python
df
```

```python
df["experience"]=df["experience"].fillna("Zero")
```

```python
df["experience"]=df["experience"].apply(w2n.word_to_num)
```

```python
df["test_score(out of 10)"]=df["test_score(out of 10)"].fillna(df["test_score(out of 10)"].median())
```

```python
df
```

```python
df.columns
```

```python
X=df.iloc[:,0:3]
y=df.iloc[:,-1]
```

```python
X
```

```python
y
```

```python
lr=LinearRegression()
lr.fit(X,y)
```

```python
y_pred=lr.predict(X)
y_pred
```

```python
df.info()
```

```python
exp=float(input("Enter years of experience:"))
test_score=float(input("Enter test score:"))
interview_score=float(input("Enter interview score:"))
user_pred=lr.predict([[exp,test_score, interview_score]])[0]
print("Predicted salary($):", user_pred)
```

```python
plt.plot(df.index, y, label='Actual Salary', marker='o')
plt.plot(df.index, y_pred, label='Predicted Salary', marker='x')
plt.scatter(len(df), user_pred, color='red', label='Your predicted salary', s=100, marker='*')


plt.xlabel("Data point index")
plt.ylabel("Salary")
plt.title("ACtual vs predicted graph")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
mae=mean_absolute_error(y,y_pred)
mse=mean_squared_error(y,y_pred)
rmse=np.sqrt(mean_squared_error(y,y_pred))
r2=r2_score(y,y_pred)
print('MAE:', mae)
print('MSE:', mse)
print('RMAE:', rmse)
print('r2:', r2)
```
