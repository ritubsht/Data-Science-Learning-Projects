```python
X=df[['Avg. Session Length', 'Time on App',
       'Time on Website', 'Length of Membership']]
y=df['Yearly Amount Spent']
```

```python
X1=df[['Time on App', 'Time on Website', 'Length of Membership']]
y1=df['Yearly Amount Spent']
```

```python
X.shape
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
X1_train, X1_test, y1_train, y1_test = train_test_split(X1, y1, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm=LinearRegression()
```

```python
lm.fit(X_train,y_train)
```

```python
lm.fit(X1_train,y1_train)
```

```python
lm.coef_
```

```python
cdf= pd.DataFrame(lm.coef_, X.columns, columns=['Coef'])
print(cdf)
```

```python
cdf= pd.DataFrame(lm.coef_, X1.columns, columns=['Coef'])
print(cdf)
```

```python
prediction=lm.predict(X_test)
```

```python
plt.figure(figsize=(6,5))
sns.scatterplot(x=y_test, y= prediction, color='r', label='Predicted')
sns.scatterplot(x=y_test, y=y_test, color='b', label='Actual')


plt.xlabel("ACtual values")
plt.ylabel("predicted values")
plt.title("GRAPH")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(6,5))
sns.scatterplot(x=y1_test, y= prediction1, color='r', label='Predicted')
sns.scatterplot(x=y1_test, y=y1_test, color='b', label='Actual')


plt.xlabel("ACtual values")
plt.ylabel("predicted values")
plt.title("GRAPH")
plt.legend()
plt.show()
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
mae=mean_absolute_error(y_test,prediction)
mse=mean_squared_error(y_test,prediction)
rmse=np.sqrt(mse)
print('MAE:', mae)
print('MSE:', mse)
print('RMAE:', rmse)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
mae=mean_absolute_error(y1_test,prediction1)
mse=mean_squared_error(y1_test,prediction1)
rmse=np.sqrt(mse)
print('MAE:', mae)
print('MSE:', mse)
print('RMAE:', rmse)
```
