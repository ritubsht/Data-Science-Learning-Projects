```python
import warnings
warnings.filterwarnings("ignore")
import nltk
import string
nltk.download('punkt_tab')
from nltk.tokenize import sent_tokenize, word_tokenize
```

```python
text="""Natural  Language Processing is fun to learn.
It helps computers understand human language !
let's explore tokeization today"""
```

```python
text_lower=text.lower()
print("After Lowering:", text_lower)
```

```python
sentences =sent_tokenize(text_lower)
print ("After sentence tokenization\n", sentences)
```

```python
words =word_tokenize(text_lower)
print ("After word tokenization\n", words)
```

```python
for i,s in enumerate(sentences,1):
    print(f"\nsentences {i}:{s}")
    print("tokens:",word_tokenize(s))
```

```python
print(string.punctuation)
print("""*****************************************************\n These are all the punctuatoions characters that python recognizes by default, we need to remove these characters.""")
```

```python
clean_words=[]
#loop through each token
for i in words:
    #keep only words that are not punctuations.
    if i not in string.punctuation:
        clean_words.append(i)    
print("\noriginal words:\n",words)
print("\nAfter removing punctuation:",clean_words)
```

```python
len(clean_words)
```
