```python
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()
words = {"Running","Runs","Easily","Studies","Farily","Change"}
for i in words:
  print(i,"<------->",stemmer.stem(i))
```

```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer,WordNetLemmatizer
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
t =["I love NLP ! Its amazing to analyze 1000s of messages, emails, and tweets!!!",
    "Cleaning text data isn't easy   we remove #hashtags, numbers (123), and special chars like @ or $."]
```

```python
len(t)
```

```python
clean_text =[]
for i in t:
  clean_text.append(i.lower())
  print(clean_text)
```

```python
# removing  emojis and special charate and number
def clean_special(text):
  text = re.sub(r'[^a-zA-Z\s]',"",text)
  return text
```

```python
cleaned_text = [clean_special(t) for t in clean_text ]
```

```python
cleaned_text
```

```python
#Tokenize

tokenized_text = []
for i in cleaned_text:
  tokenized_text.append(word_tokenize(i))

```

```python
tokenized_text
```

```python
stop_words = set(stopwords.words('english'))
new_text = []
for i in tokenized_text :
  for x in i:
    if x not in stop_words:
      new_text.append(x)
```

```python
#stemming
stemmer = PorterStemmer()
for i in new_text:
  print(i,"<---------->",stemmer.stem(i))
```

```python
lem = WordNetLemmatizer()
for i in new_text:
  print(i,"<---------->",lem.lemmatize(i))
```
