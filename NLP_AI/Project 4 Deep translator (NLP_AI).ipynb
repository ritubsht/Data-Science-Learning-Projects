```python
from deep_translator import GoogleTranslator
```

```python
text="Ritu Bisht"
```

```python
GoogleTranslator(source='en',target="japanese").translate(text)
```

```python
def translate(text,target):
  trans_text = GoogleTranslator(target=target).translate(text)
  return trans_text
```

```python
while True:
    print("Language Translation tool")
    input_text=input("Enter your text")
    if input_text == 'exit':
        break
    target_lang= input("To tanslate lang.(en,hi,ur,ta,pun)")
    trans_text=translate(input_text, target_lang)
    print("Translation:", trans_text)
```
