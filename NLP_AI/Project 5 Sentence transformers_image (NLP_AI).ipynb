```python
from sentence_transformers import SentenceTransformer
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
#load the model
model=SentenceTransformer("clip-Vit-B-32")
```

```python
image1='/home/ritu-bisht/Desktop/RITU BISHT/JN/Image1.jpeg'
```

```python
#Your data
text_description = [
    "Two dogs in the snow",
    "A cat on a table",
    "A picture of London at night",
    "A happy dog playing fetch",
    " A dog sleeping on a couch",
    "Two cats fighting in the garden"
]
```

```python
def display_image_and_compare(image_path, text_description):
  #load the image
  img = Image.open(image_path)

  #Encode text descriptions and image
  text_emb = model.encode(text_description)
  image_emb = model.encode(img)

  #calculate cosine similarity
  similarity_score = np.dot(text_emb, image_emb.T)

  #find the best match
  best_match_index = np.argmax(similarity_score)
  best_match_desc = text_description[best_match_index]

  #Print best match
  print(f"\nBest Match for the Image: '{best_match_desc}' with a score of {similarity_score[best_match_index]:.4f}")

  #Display the image
  plt.figure(figsize=(6,6))    #Make it larger
  plt.imshow(img)              #display image data
  plt.axis('off')              #hide axis
  plt.title("Image for similarity comparison", fontsize=14)
  plt.show()                   #show the image
```

```python
display_image_and_compare(image1, text_description)
```
