```python
def emotion_detection(text):
    emotion = create_analyzer(task='emotion',lang = "en")
    result = emotion.predict(text)
    return result.output
```

```python
emotion_detection("i am scared of leaerning maths")
```

```python
next(iter(result3.probas.items()))
```

```python
def hate_speech_detect(text):
    hatespeech = create_analyzer(task = "hate_speech",lang="en")
    result3 = hatespeech.predict(text)
    return next(iter(result3.probas.items()))
```

```python
hate_speech_detect("kill them")
```

```python
text = """I woke up this morning feeling incredibly happy and grateful for the beautiful weather outside.
The sun was shining, birds were singing, and there was a gentle breeze in the air. It felt like the
perfect day to go for a long walk in the park."""

emo = emotion_detection(text)
hate = hate_speech_detect(text)

print("Emotion :", emo)
print("Speech :", hate)
```

```python
text2 = """You are the dumbest person I have ever met.  
Your ideas are trash, and nobody wants to hear you speak.
"""

emo = emotion_detection(text2)
hate = hate_speech_detect(text2)

print("Emotion :", emo)
print("Speech :", hate)
```

```python
text3 ="""Get lost, you useless idiot.  
People like you ruin everything with your stupidity.
"""

emo = emotion_detection(text3)
hate = hate_speech_detect(text3)

print("Emotion :", emo)
print("Speech :", hate)
```

```python
texts = [
    """I just received the best news today! My project got approved, and I can’t stop smiling. 
    Everything feels so positive, and I’m truly grateful.""",

    """You are the dumbest person I have ever met. Your ideas are trash and nobody wants to hear you speak."""
]

for t in texts:
    emo = emotion_detection(t)
    hate = hate_speech_detect(t)
    print("Text:", t)
    print("Emotion:", emo)
    print("Hate Speech:", hate)
    print("------")
```
